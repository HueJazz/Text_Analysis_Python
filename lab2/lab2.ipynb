```python
#1.Зчитати файл text5
text_file = open("assets/text5.txt")
text = text_file.read()
text
```

```python
import nltk
nltk.download('all')
```

```python
#а)  Порахувати  кількість слів  в  останньому реченні(не  враховуючи  знаки  пунктуації  та  інші  символи);
from nltk.tokenize import sent_tokenize
sents = sent_tokenize(text, language='english')
print(len(sents[-1].split()))
print(len(nltk.WhitespaceTokenizer().tokenize(sents[-1])))
```

```python
#б)  видалити стоп-слова;
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stoplist = stopwords.words('english')
punctuation = [',', '.']
cleanwordlist = [word for word in word_tokenize(text.lower()) if word not in stoplist if word not in punctuation]
cleanwordlist
```

```python
#в)  знайти 9 слів, які зустрічаються найчастіше.
cleanwordlist = [word for word in word_tokenize(text.lower()) if word not in punctuation]
freq_list = nltk.FreqDist(cleanwordlist)
print(freq_list)
freq_list = [x for x in list(freq_list)[0:9]]
print(freq_list)
```

```python
#2.Використати корпус Reuters, другий  текст  категорії cotton.
#а)  Вивести речення через одне;
from nltk.corpus import reuters
text2 = reuters.sents(reuters.fileids('cotton')[1])
text2 = [' '.join(sentence_token) for sentence_token in text2]
print(text2)
sentences = [text2[i] for i in range(len(text2)) if i%2==1]
print(sentences)
```

```python
#б)  Вивести 10 дієслів, що зустрічаються найчастіше.
cleanwordlist = [word.lower() for word in reuters.words(reuters.fileids('cotton')[1])]
verb_list = ['VBD', 'VBN', 'VBG', 'VBP', 'VB', 'VBZ']
verbs = [(word, tag) for word, tag in nltk.pos_tag(cleanwordlist) if any(verb_tag in tag for verb_tag in verb_list)]
print(verbs)
verbs_freq = nltk.FreqDist([word for word, tag in verbs])
print(verbs_freq)
verbs_freq = [x for x in list(verbs_freq)[0:10]]
print(verbs_freq)
```
